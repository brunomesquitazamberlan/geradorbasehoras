```python
import streamlit as st
import pandas as pd


```

```python
def transforma_csv(input_df):
    # Função de exemplo que transforma o DataFrame de entrada
    input = pd.read_csv(input_df)
    
    df1 = input.copy()
    df1_result = df1.to_csv("df1.csv")
    df2 = input.copy()
    df2_result = df2.to_csv("df2.csv")

    # Exemplos de transformações
    return df1_result, df2_result
```

```python
def adjusted_transforma_csv(input_df):
    # Função de exemplo que transforma o DataFrame de entrada
    input_dataframe = pd.read_csv(input_df)
    
    df1 = input_dataframe.copy()
    df2 = input_dataframe.copy()
    
    # Exemplos de transformações
    return df1, df2
```
